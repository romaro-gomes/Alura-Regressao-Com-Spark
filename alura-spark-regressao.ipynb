# Preparando os dados para o modelo

## Variáveis Dummy

```python
dataset\
.groupBy("customerID")\
.pivot('unit')\
.agg(f.lit(1))\
.na\
.fill(0)\
.show()
```

```python
unit=dataset\
.groupBy("customerID")\
.pivot('unit')\
.agg(f.lit(1))\
.na\
.fill(0)
```

```python
zone=dataset\
.groupBy("customerID")\
.pivot('zone')\
.agg(f.lit(1))\
.na\
.fill(0)
```

```python
## Unindo doa datasets
dataset=dataset\
.join(unit,'customerID',how='inner')\
.join(zone,'customerID',how='inner')
```

```python
# Vetorizaçõ dos dados
from pyspark.ml.feature import VectorAssembler
```

```python
# O spark espera que a varial predita tenha o nome label
dataset = dataset.withColumnRenamed('price','label')
```

```python
# Selecionando as variaveis de interesse do modelo
X = [
    'bathrooms',
    'bedrooms',
    'floors',
    'parkingSpaces',
    'suites',
    'unitFloor',
    'unitsOnTheFloor',
    'usableAreas',
    'condo',
    'iptu',
    'Apartamento',
    'Casa',
    'Outros',
    'Zona Central',
    'Zona Norte',
    'Zona Oeste',
    'Zona Sul'
]


```

```python
assembler= VectorAssembler(inputCols=X,outputCol='features') # opyspark espera receber featurs como ariavel preditora
```

```python
dataset_prep = assembler.transform(dataset).select('features','label')
```

```python
dataset_prep.show(10, truncate=False)
```

# Técnicas de Otimização

## Cross-Validation

```python
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
```

### Decision Tree

```python
dtr=DecisionTreeRegressor()
```

```python
# Parametros para treino
grid=ParamGridBuilder()\
.addGrid(dtr.maxDepth, [2,5,10])\
.addGrid(dtr.maxBins, [10,32,45])\
.build()
```

```python
evaluator=RegressionEvaluator()
```

```python
dtr_cv= CrossValidator(
    estimator=dtr,
    estimatorParamMaps=grid,
    evaluator=evaluator,
    numFolds=3,
    seed = 101)
```

```python
modelo_dtr_cv=dtr_cv.fit(treino)
```

```python
previsoes_dtr_cv_teste= modelo_dtr_cv.transform(teste)
```

```python
print('Decision Tree Regression')
print("="*30)
print("Sem Cross Validation")
print("="*30)
print("R²: %f" % evaluator.evaluate(previsoes_dtr_teste, {evaluator.metricName: "r2"}))
print("RMSE: %f" % evaluator.evaluate(previsoes_dtr_teste, {evaluator.metricName: "rmse"}))
print("")
print("="*30)
print("Com Cross Validation")
print("="*30)
print("R²: %f" % evaluator.evaluate(previsoes_dtr_cv_teste, {evaluator.metricName: "r2"}))
print("RMSE: %f" % evaluator.evaluate(previsoes_dtr_cv_teste, {evaluator.metricName: "rmse"}))
```

### Random Forest
